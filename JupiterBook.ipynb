### ANALISIS DE LA BASE DE DATOS

**Variables categ칩ricas**
- **Potability**: Potabilidad del agua.
    - **Descripci칩n**: Indica si el agua es potable con valores 1 (potable) y 0 (no potable).

**Variables continuas**
- **pH**: Mide el nivel de acidez o alcalinidad del agua.
    - **Descripci칩n**: En el Data Frame, podemos ver como el pH en las muestras potables, es del rango establecido como "saludable", que va desde 6.5 hasta 8.5
- **Hardness**: Minerales disueltos en el agua.
    - **Descripci칩n**: Las mismas son encontradas en las muestras por el roce con las piedras, arenas, plantas...El tiempo que el agua esta en contacto con estas sales/minerales, determina su dureza.
- **Solids**: S칩lidos organicos/no organicos totales disueltos en el agua.
    - **Descripci칩n**: Estos minerales producen efectos de sabor/coloracion en el agua. Un alto nivel de solidos, indica una muestra demasiado mineralizada. Lo deseable se encuentra entre 500mg/l y 1000mg/l.
- **Chloramines**: Desinfectantes (cloros).
    - **Descripci칩n**: La cloraci칩n tiene como objetivo principal eliminar o inactivar microorganismos pat칩genos, como bacterias, virus y par치sitos, que podr칤an estar presentes en el agua y que podr칤an causar enfermedades si son ingeridos. Los limites saludables son de hasta 4mg/l.
- **Sulfate**: Sustancias encontradas en minerales, piedras, plantas...
    - **Descripci칩n**: Los sulfatos son una mezcla de ox칤geno y azufre y son partes de las sustancias existentes en algunas formaciones de rocas y suelos que incorporan agua subterr치nea. El mineral se convierte gradualmente en una soluci칩n y se libera al agua subterr치nea.
- **Conductivity**: Conductividad de la muestra.
    - **Descripci칩n**: Las sales disueltas en agua se descomponen en iones cargados positivamente y negativamente. La conductividad se define como la capacidad del agua para conducir una corriente el칠ctrica a trav칠z de los iones disueltos. Los iones m치s positivos son sodio (Na+), calcio (Ca+2), potasio (K+) y magnesio (Mg+2). No se deberia exceder de los 400풮S/cm
- **Organic_carbon**: Representa el contenido de carbono org치nico en el agua.
    - **Descripci칩n**: El carbono org치nico total es la medida de todo el carbono que contiene la materia org치nica disuelta o particulada presente en la muestra de agua. El origen de esta materia org치nica es, principalmente, sustancias vegetales o animales, contaminaciones que se producen en el sistema de distribuci칩n. Su presencia por encima de los valores establecidos, al reaccionar con los agentes desinfectantes utilizados en la potabilizaci칩n, como el cloro, pueden dar lugar a subproductos de la desinfecci칩n nocivos para la salud. Lo recomendable se encuentra en < 2mg/l cuando el agua es potable, y < 4mg/l cuando el agua esta en tratamiento
- **Trihalomethanes**: Concentraci칩n de trihalometanos en el agua.
    - **Descripci칩n**: Los trihalometanos son unos subproductos de la desinfecci칩n que se forman cuando se emplea el cloro como desinfectante. Los trihalometanos que se encuentran en el agua de consumo humano son el cloroformo, el bromodiclorometano, el dibromoclorometano y el bromoformo. Hasta 80 ppm, se consideran seguros.
- **Turbidity**: Grado de transparencia que pierde el agua.
    - **Descripci칩n**: La turbidez es una medida de la cantidad de part칤culas en suspensi칩n en el agua. Cuanto mayor sea la cantidad de s칩lidos suspendidos en el l칤quido, como las algas, sedimentos,materia org치nica y los contaminantes, mayor ser치 el grado de turbidez.

### RED NEURONAL

#### Arquitectura de la red

- Capas
    - La red dispondria de 3 capas: 1 de entrada, 1 capa oculta y 1 de salida.

- Neuronas
    - En la capa de entrada habra tantas neuronas como columnas (datos de entrada), en la capa oculta, tambien habra 9 neuronas (ni idea, preguntar) y en la ultima 1 sola, ya que es 1 si es potable, y 0 si no lo es.

- Funciones de Activacion
    - ReLu en las ocultas, Sigmoide en la de salida

#### Separacion Train/Test

Para empezar con la red, separamos de una vez los datos en conjuntos de entrenamiento y testeo, para despues poder ir repitiendo el entrenamiento siempre con los mismos datos.

```python
# Extraigo las columnas de entrada
inputs = df.iloc[:, 0:9].values
outputs = df.iloc[:, -1].values

# Conjuntos de entrenamiento y prueba
x_train, x_test, y_train, y_test = train_test_split(inputs, outputs, test_size=1/3)

#shape retorna una tupla con las dimensiones de la matriz = (filas, columnas).
# por lo que shape[0], nos retorna las filas de la matriz.
n_train = x_train.shape[0] # numero de registros de entrenamiento
# Esto no se usa para nada?
n_test = x_test.shape[0] # numero de registros de testeo
print(n_train)
print(n_test)
```

Construimos la red neuronal con pesos y sesgos aleatorios

```python
def red_neuronal(L, epochs):
    np.random.seed(0)
    # pesos
    # Preguntar por esto, nunca lo termine de entender, aca que estoy haciendo? 6 neuronas en la capa oculta es lo que supongo, <-
    # pero capaz esta mal.
    w_hidden_1 = np.random.rand(9,9) * 2 - 1
    w_output_1 = np.random.rand(1,9) * 2 - 1

    # sesgos
    b_hidden = np.random.rand(9,1) * 2 - 1
    b_output = np.random.rand(1,1) * 2 - 1

    # Funciones de Activacion
    relu = lambda x: np.maximum(x, 0)
    leaky_relu = lambda x: x if np.greater_equal(x,0) else x * 0.01
    sigmoide = lambda x: 1 / (1 + np.exp(-x))

    def f_prop(X):
        z1 = w_hidden_1 @ X + b_hidden
        a1 = leaky_relu(z1)
        z2 = w_output_1 @ a1 + b_output
        a2 = sigmoide(z2)
        return z1, a1, z2, a2

    # Derivadas de las funciones de activaci칩n
    d_relu = lambda x: x > 0
    d_leaky_relu = lambda x: 1 if np.greater_equal(x,0) else 0.01
    d_sigmoide = lambda x: np.exp(-x) / (1 + np.exp(-x)) ** 2

    # Devuelve pendientes para pesos y sesgos
    # usando la regla de la cadena
    def b_prop(z1, a1, z2, a2, X, Y):
        dC_dA2 = 2 * a2 - 2 * Y
        dA2_dZ2 = d_sigmoide(z2)
        dZ2_dA1 = w_output_1
        dZ2_dW2 = a1
        dZ2_dB2 = 1
        dA1_dZ1 = d_leaky_relu(z1)
        dZ1_dW1 = X
        dZ1_dB1 = 1

        dC_dW2 = dC_dA2 @ dA2_dZ2 @ dZ2_dW2.T
        dC_dB2 = dC_dA2 @ dA2_dZ2 * dZ2_dB2
        dC_dA1 = dC_dA2 @ dA2_dZ2 @ dZ2_dA1
        dC_dW1 = dC_dA1 @ dA1_dZ1 @ dZ1_dW1.T
        dC_dB1 = dC_dA1 @ dA1_dZ1 * dZ1_dB1

        return dC_dW1, dC_dB1, dC_dW2, dC_dB2

    # Para el grafico de las funciones
    accuracy_train_l = []
    accuracy_test_l = []

    # Ejecutar descenso de gradiente
    for i in range(epochs):
      # seleccionar aleatoriamente uno de los datos de entrenamiento
      idx = np.random.choice(n_train, 1, replace=False)
      X_sample = x_train[idx].transpose()
      Y_sample = y_train[idx]

      # pasar datos seleccionados aleatoriamente a trav칠s de la red neuronal
      Z1, A1, Z2, A2 = f_prop(X_sample)

      # distribuir error a trav칠s de la retropropagaci칩n
      # y devolver pendientes para pesos y sesgos
      dW1, dB1, dW2, dB2 = b_prop(Z1, A1, Z2, A2, X_sample, Y_sample)

      # actualizar pesos y sesgos
      w_hidden_1 -= L * dW1
      b_hidden -= L * dB1
      w_output_1 -= L * dW2
      b_output -= L * dB2

      # Accuracy
      # Test
      test_predictions = f_prop(x_test.transpose())[3]
      test_comparisons = np.equal((test_predictions >= .5).flatten().astype(int), np.array(y_test >= .5).astype(int))
      accuracy_test = sum(test_comparisons.astype(int) / x_test.shape[0])
      accuracy_test_l.append(accuracy_test)
      
      # Train
      train_predictions = f_prop(x_train.transpose())[3]
      train_comparisons = np.equal((train_predictions >= .5).flatten().astype(int), np.array(y_train >= .5).astype(int))
      accuracy_train = sum(train_comparisons.astype(int) / x_train.shape[0])
      accuracy_train_l.append(accuracy_train)

    print("TEST ACCURACY: ", accuracy_test)
    print("TRAIN ACCURACY: ", accuracy_train)
    print(f'Epochs: {epochs}')
    print(f'Paso: {L}')
    grafico_acc(L=L, train_l=accuracy_train_l, test_l=accuracy_test_l)
```
