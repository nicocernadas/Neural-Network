### Regresi√≥n lineal b√°sica con Scikit-Learn

Vamos a hacer una regresi√≥n lineal utilizando *Python*, comenzando con **scikit-learn** para facilitar el trabajo.

El ejemplo muestra c√≥mo utilizamos **scikit-learn** para realizar una regresi√≥n lineal b√°sica y no validada en la muestra de 10 perros. Importamos estos datos utilizando *Pandas*, los convertimos en arrays de *NumPy*, realizamos la regresi√≥n lineal usando *scikit-learn* y utilizamos *Plotly* para mostrarla en un gr√°fico.

````python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Importar puntos
df = pd.read_csv('https://bit.ly/3goOAnt', delimiter=",")

# Extraer variables de entrada (todas las filas, todas las columnas menos la √∫ltima)
X = df.values[:, :-1]

print(X)

# Extraer columna de salida (todas las filas, √∫ltima columna)
Y = df.values[:, -1]

print(Y)

# Ajustar una l√≠nea a los puntos
fit = LinearRegression().fit(X, Y)

m = fit.coef_.flatten()
b = fit.intercept_.flatten()
print(f"m = {m}")
print(f"b = {b}")

# mostrar en gr√°fico
plt.plot(X, Y, 'o')  # diagrama de dispersi√≥n
plt.plot(X, m*X+b)  # l√≠nea
plt.show()
````

## Residuos y Errores Cuadrados


¬øC√≥mo herramientas estad√≠sticas como *scikit-learn* determinan una l√≠nea que se ajusta a estos puntos?  
Se reduce a dos preguntas fundamentales para el entrenamiento en aprendizaje autom√°tico:

* ¬øQu√© define un "mejor ajuste"?
* ¬øC√≥mo llegamos a ese "mejor ajuste"?

La primera pregunta tiene una respuesta bastante establecida: minimizamos los cuadrados, o m√°s espec√≠ficamente, la suma de los residuos al cuadrado. Desglosemos esto.

Dibuja cualquier l√≠nea a trav√©s de los puntos. El residuo es la diferencia num√©rica entre la l√≠nea y los puntos, como se muestra a continuaci√≥n:  



Los puntos que est√°n por encima de la l√≠nea tendr√°n un residuo positivo, y los puntos que est√°n por debajo de la l√≠nea tendr√°n un residuo negativo. En otras palabras, es la diferencia sustra√≠da entre los valores predichos de *y* (derivados de la l√≠nea) y los valores reales de *y* (que provienen de los datos). Otro nombre para los residuos son errores, porque reflejan cu√°n equivocada est√° nuestra l√≠nea al predecir los datos.

Calculemos estas diferencias entre estos 10 puntos y la l√≠nea *y = 1.93939x + 4.73333* y los residuos para cada punto:

````python
import pandas as pd

# Importar puntos
points = pd.read_csv('https://bit.ly/3goOAnt', delimiter=",").itertuples()

# Prueba con una l√≠nea dada
m = 1.93939
b = 4.73333

# Calcular los residuos
for p in points:
    # print(p)
    y_actual = p.y
    y_predict = m * p.x + b
    residual = y_actual - y_predict
    print(residual)
````

Si estamos ajustando una l√≠nea recta a trav√©s de nuestros 10 puntos de datos, probablemente querremos minimizar estos residuos en total para que haya la menor distancia posible entre la l√≠nea y los puntos. Pero, ¬øc√≥mo medimos el "total"?  

El mejor enfoque es tomar la suma de cuadrados, que simplemente eleva al cuadrado cada residuo, o multiplica cada residuo por s√≠ mismo, y los suma.  

Tomamos cada valor real de *y* y le restamos el valor predicho de *y* tomado de la l√≠nea, luego cuadramos y sumamos todas esas diferencias.

**¬øPOR QU√â NO VALORES ABSOLUTOS?**
Te podr√≠as preguntar por qu√© tenemos que elevar al cuadrado los residuos antes de sumarlos. ¬øPor qu√© no simplemente sumarlos sin elevar al cuadrado? Eso no funcionar√≠a porque los negativos cancelar√≠an a los positivos. ¬øY si sumamos los valores absolutos, donde convertimos todos los valores negativos en positivos? Eso suena prometedor, pero los valores absolutos son matem√°ticamente inconvenientes. M√°s espec√≠ficamente, los valores absolutos no funcionan bien con las derivadas del c√°lculo que vamos a usar m√°s adelante para el descenso del gradiente. Es por esto que elegimos los residuos al cuadrado como nuestra forma de totalizar la p√©rdida.

Una manera visual de pensar en esto se muestra a continuaci√≥n, donde superponemos un cuadrado en cada residuo y cada lado es la longitud del residuo. Sumamos el √°rea de todos estos cuadrados, y m√°s adel√°nte aprenderemos c√≥mo encontrar la suma m√≠nima que podemos lograr identificando el mejor *m* y *b*.  



Vamos a modificar nuestro c√≥digo en el Ejemplo 5-4 para encontrar la suma de cuadrados.

````python
import pandas as pd

# Importar puntos
points = pd.read_csv("https://bit.ly/2KF29Bd").itertuples()

# Prueba con una l√≠nea dada
m = 1.93939
b = 4.73333

sum_of_squares = 0.0

# calcular la suma de cuadrados
for p in points:
    y_actual = p.y
    y_predict = m*p.x + b
    residual_squared = (y_predict - y_actual)**2
    sum_of_squares += residual_squared

print(f"suma de cuadrados = {sum_of_squares}")
# suma de cuadrados = 28.096969704500005
````

## Descenso de Gradiente

El descenso de gradiente es una t√©cnica de optimizaci√≥n que utiliza derivadas e iteraciones para minimizar/maximizar un conjunto de par√°metros contra un objetivo. Para aprender sobre el descenso de gradiente, hagamos un peque√±o experimento mental y luego apliqu√©moslo en un ejemplo sencillo.

### Un Experimento Mental sobre el Descenso de Gradiente

El descenso de gradiente es una t√©cnica para optimizar que, usando derivadas e iteraciones, busca minimizar/maximizar un conjunto de par√°metros frente a un objetivo. Imaginate que est√°s en una monta√±a de noche con una linterna, buscando el punto m√°s bajo. Este m√©todo, donde avanz√°s hacia donde la pendiente baja, se llama descenso de gradiente.

En el aprendizaje autom√°tico, queremos minimizar nuestra p√©rdida. Para eso el descenso de gradiente tiene una caracter√≠stica atractiva: la derivada parcial es esa linterna, permiti√©ndonos ver las pendientes para cada par√°metro (en este caso *m* y *b*). Avanzamos en direcciones para donde la pendiente desciende.  

Damos pasos m√°s grandes para pendientes m√°s pronunciadas y pasos m√°s peque√±os para pendientes menos pronunciadas. Podemos simplemente calcular la longitud de este paso tomando una fracci√≥n de la pendiente. Esta fracci√≥n se conoce como nuestra **tasa de aprendizaje**. Cuanto mayor sea la tasa de aprendizaje, m√°s r√°pido avanzar√° a costa de la precisi√≥n. Pero cuanto menor sea la tasa de aprendizaje, m√°s tiempo tomar√° entrenar y requerir√° m√°s iteraciones.

Decidir una tasa de aprendizaje es como elegir entre una hormiga, un humano o un gigante para bajar por la pendiente. Una hormiga (tasa de aprendizaje peque√±a) tomar√° pasos diminutos y tomar√° un tiempo inaceptablemente largo para llegar al fondo, pero lo har√° con precisi√≥n. Un gigante (tasa de aprendizaje grande) puede seguir pasando por encima del m√≠nimo al punto de que nunca pueda alcanzarlo, sin importar cu√°ntos pasos d√©. El humano (tasa de aprendizaje moderada) probablemente tenga el tama√±o de paso m√°s equilibrado, teniendo el justo equilibrio entre velocidad y precisi√≥n al llegar al m√≠nimo.

### Caminemos Antes de Correr

Aunque podr√≠amos resolver esto algebraicamente, usemos el descenso de gradiente para hacerlo.

## Descenso de Gradiente y Regresi√≥n Lineal

Ahora, te podr√≠as estar preguntando c√≥mo usamos esto para la regresi√≥n lineal. Es la misma idea, pero nuestras "variables" son *m* y *b*, no *x*.  
En la regresi√≥n lineal ya conocemos los valores de *x* e *y* porque son nuestros datos de entrenamiento. Lo que necesitamos solucionar son los par√°metros *m* y *b* para encontrar la mejor l√≠nea de ajuste.

Usando descenso de gradiente para encontrar el m√≠nimo de una par√°bola

Se proporciona un c√≥digo que utiliza el descenso de gradiente para encontrar el punto m√≠nimo de una funci√≥n parab√≥lica, ajustando el valor de *x* mediante iteraciones.

````python
import random

def f(x):
    return (x - 3) ** 2 + 4

def dx_f(x):
    return 2 * (x - 3)

L = 0.001  # La tasa de aprendizaje
iterations = 100_000  # La cantidad de iteraciones
x = random.randint(-15, 15)  # Empezamos en un x al azar

for i in range(iterations):
    d_x = dx_f(x)  # La pendiente
    x -= L * d_x  # Actualizamos x

print(x, f(x))  # Deber√≠a acercarse a 3, 4
````

Este c√≥digo va reduciendo la diferencia entre nuestro *x* y el punto m√°s bajo de la funci√≥n, usando el descenso de gradiente.

### Realizando descenso de gradiente para una regresi√≥n lineal

Muestra c√≥mo implementar el descenso de gradiente para ajustar los par√°metros *m* y *b* de una regresi√≥n lineal, utilizando una tasa de aprendizaje y un n√∫mero espec√≠fico de iteraciones para minimizar la suma de cuadrados de los residuos.

````python
import pandas as pd

points = list(pd.read_csv("https://bit.ly/2KF29Bd").itertuples())
m = b = 0.0  # Iniciamos m y b en 0
L = 0.001  # La tasa de aprendizaje
iterations = 100_000  # La cantidad de iteraciones

for i in range(iterations):
    D_m = sum(2 * p.x * ((m * p.x + b) - p.y) for p in points)  # Derivada parcial respecto a m
    D_b = sum(2 * ((m * p.x + b) - p.y) for p in points)  # Derivada parcial respecto a b
    m -= L * D_m  # Actualizamos m
    b -= L * D_b  # Actualizamos b

print(f"y = {m}x + {b}")  # Mostramos la ecuaci√≥n de la l√≠nea
````

### Calculando derivadas parciales para *m* y *b* con SymPy
Presenta c√≥mo usar *SymPy* para calcular autom√°ticamente las derivadas parciales necesarias para el descenso de gradiente en el contexto de una regresi√≥n lineal.

````python
from sympy import symbols, Function, Sum, diff

m, b, i, n = symbols('m b i n')
x, y = symbols('x y', cls=Function)
sum_of_squares = Sum((m*x(i) + b - y(i)) ** 2, (i, 0, n))

d_m = diff(sum_of_squares, m)
d_b = diff(sum_of_squares, b)
print(d_m)
print(d_b)
````


Si quer√©s aplicar nuestro dataset y ejecutar una regresi√≥n lineal usando el descenso de gradiente, tendr√°s que realizar unos pasos adicionales como se muestra en el ejemplo que viene. Necesitaremos sustituir los valores de *n*, *x(i)* y *y(i)*, iterando todos nuestros puntos de datos para las funciones derivadas.  

Esto nos dejar√° solo con las variables *m* y *b*, para las cuales buscaremos los valores √≥ptimos usando el descenso de gradiente.

````python
import pandas as pd
from sympy import *

# Importar puntos desde CSV
points = list(pd.read_csv("https://bit.ly/2KF29Bd").itertuples())

m, b, i, n = symbols('m b i n')
x, y = symbols('x y', cls=Function)

sum_of_squares = Sum((m*x(i) + b - y(i)) ** 2, (i, 0, n))

d_m = diff(sum_of_squares, m) \
    .subs(n, len(points) - 1).doit() \
    .replace(x, lambda i: points[i].x) \
    .replace(y, lambda i: points[i].y)

d_b = diff(sum_of_squares, b) \
    .subs(n, len(points) - 1).doit() \
    .replace(x, lambda i: points[i].x) \
    .replace(y, lambda i: points[i].y)

# compilar usando lambdify para una computaci√≥n m√°s r√°pida
d_m = lambdify([m, b], d_m)
d_b = lambdify([m, b], d_b)

# Construyendo el modelo
m = 0.0
b = 0.0

# La tasa de aprendizaje
L = .001

# La cantidad de iteraciones
iterations = 100_000

# Realizar Descenso de Gradiente
for i in range(iterations):
    # actualizar m y b
    m -= d_m(m,b) * L
    b -= d_b(m,b) * L

print(f"y = {m}x + {b}")
````

Es una buena idea llamar a lambdify() en ambas funciones de derivadas parciales para convertirlas de SymPy a una funci√≥n de Python optimizada. Esto har√° que los c√°lculos se realicen mucho m√°s r√°pidamente cuando hagamos descenso de gradiente. Las funciones de Python resultantes est√°n respaldadas por NumPy, SciPy o cualquier biblioteca num√©rica que SymPy detecte que tienes disponible. Despu√©s de eso, podemos realizar el descenso de gradiente.

Finalmente, si tienes curiosidad sobre c√≥mo se ve la funci√≥n de p√©rdida para esta regresi√≥n lineal simple, el ejemplo que viene muestra el c√≥digo de SymPy que inserta los valores de *x*, *y* y *n* en nuestra funci√≥n de p√©rdida y luego grafica *m* y *b* como las variables de entrada.  

Nuestro algoritmo de descenso de gradiente nos lleva al punto m√°s bajo en este paisaje de p√©rdida mostrado en la imagen.

````python
from sympy import *
from sympy.plotting import plot3d
import pandas as pd

points = list(pd.read_csv("https://bit.ly/2KF29Bd").itertuples())
m, b, i, n = symbols('m b i n')
x, y = symbols('x y', cls=Function)

sum_of_squares = Sum((m*x(i) + b - y(i)) ** 2, (i, 0, n)) \
    .subs(n, len(points) - 1).doit() \
    .replace(x, lambda i: points[i].x) \
    .replace(y, lambda i: points[i].y)

plot3d(sum_of_squares)
````

## Sobreajuste (overfitting) y Varianza

Pregunta: si realmente queremos minimizar la p√©rdida, es decir, reducir la suma de cuadrados a 0, ¬øqu√© har√≠amos? ¬øHay opciones adem√°s de la regresi√≥n lineal?  

Una conclusi√≥n a la que podr√≠as llegar es simplemente ajustar una curva que toque todos los puntos.  

Es decir, ¬øpor qu√© no simplemente conectar los puntos en segmentos y usar eso para hacer predicciones? ¬°Eso nos da una p√©rdida de 0!  



Che, ¬øpor qu√© nos metimos en todo ese l√≠o con la regresi√≥n lineal y no hicimos esto en su lugar?  

Bueno, record√° que nuestro objetivo a gran escala no es minimizar la suma de cuadrados sino hacer predicciones precisas sobre nuevos datos. Este modelo de conectar-los-puntos est√° severamente sobreajustado, lo que significa que se molde√≥ la regresi√≥n a los datos de entrenamiento demasiado exactamente al punto de que predecir√° mal sobre nuevos datos.  

Este simple modelo de conectar los puntos es sensible a los valores at√≠picos que est√°n lejos del resto de los puntos, lo que significa que tendr√° alta varianza en las predicciones. Aunque los puntos en este ejemplo est√°n relativamente cerca de una l√≠nea, este problema ser√° mucho peor con otros conjuntos de datos con m√°s dispersi√≥n y valores at√≠picos. ¬°Porque el sobreajuste aumenta la varianza, las predicciones van a ser muy variadas!

<div class="alert alert-block alert-warning">
<b>El Sobreajuste es Memorizaci√≥n:</b>
Cuando alguien dice que una regresi√≥n "memoriz√≥" los datos en lugar de generalizarlos, est√°n hablando de sobreajuste.

Como pod√©s adivinar, queremos encontrar generalizaciones efectivas en nuestro modelo en lugar de memorizar datos. De lo contrario, nuestra regresi√≥n simplemente se convierte en una base de datos donde buscamos valores.
</div>

Por eso en el aprendizaje autom√°tico encontrar√°s que se agrega sesgo al modelo, y la regresi√≥n lineal se considera un modelo con alto sesgo.

Esto no es lo mismo que el sesgo en los datos, sobre lo cual hablamos extensamente la clase anterior. El sesgo en un modelo significa que priorizamos un m√©todo (por ejemplo, mantener una l√≠nea recta) en lugar de doblarnos y ajustarnos exactamente a lo que dicen los datos.  

Un modelo sesgado deja cierto margen de maniobra esperando minimizar la p√©rdida en nuevos datos para obtener mejores predicciones, en lugar de minimizar la p√©rdida en los datos en los que fue entrenado.  

Supongo que podr√≠as decir que agregar sesgo a un modelo contrarresta el sobreajuste con el subajuste, o ajust√°ndose menos a los datos de entrenamiento.

Como te podr√°s imaginar, esto es un acto de equilibrio porque son dos objetivos contradictorios. En aprendizaje autom√°tico, b√°sicamente estamos diciendo: ‚ÄúQuiero ajustar una regresi√≥n a mis datos, pero no quiero ajustarla demasiado. Necesito cierto margen de maniobra para predicciones en nuevos datos que ser√°n diferentes.‚Äù

#### Regresi√≥n Lasso y Ridge

Dos variantes algo populares de la regresi√≥n lineal son la regresi√≥n lasso y la regresi√≥n ridge. La regresi√≥n ridge agrega un sesgo adicional a una regresi√≥n lineal en forma de una penalizaci√≥n, haciendo que se ajuste menos a los datos. La regresi√≥n lasso intentar√° marginalizar las variables ruidosas, lo que la hace √∫til cuando quer√©s eliminar autom√°ticamente variables que podr√≠an ser irrelevantes.  

Sin embargo, no podemos simplemente aplicar una regresi√≥n lineal a algunos datos, hacer algunas predicciones con ella y asumir que todo est√° bien. Una regresi√≥n lineal puede sobreajustarse incluso con el sesgo de una l√≠nea recta. Por lo tanto, necesitamos verificar y mitigar tanto el sobreajuste como el subajuste para encontrar el punto √≥ptimo entre los dos.  

Es decir, a menos que no haya ninguno en absoluto, en cuyo caso deber√≠as abandonar el modelo por completo.

## Descenso de Gradiente Estoc√°stico

En un contexto de aprendizaje autom√°tico, es poco probable que hagas descenso de gradiente en la pr√°ctica como lo hicimos antes, donde entrenamos con todos los datos de entrenamiento (llamado descenso de gradiente por lotes). En la pr√°ctica, es m√°s probable que realices descenso de gradiente estoc√°stico, que entrena solo con una muestra del conjunto de datos en cada iteraci√≥n. En el descenso de gradiente por mini-lotes, se utilizan m√∫ltiples muestras del conjunto de datos (por ejemplo, 10 o 100 puntos de datos) en cada iteraci√≥n.

¬øPor qu√© usar solo parte de los datos en cada iteraci√≥n? Los practicantes de aprendizaje autom√°tico citan algunos beneficios. Primero, reduce significativamente el c√°lculo, ya que cada iteraci√≥n no tiene que atravesar todo el conjunto de datos de entrenamiento, sino solo parte de √©l. El segundo beneficio es que reduce el sobreajuste. Exponer el algoritmo de entrenamiento solo a parte de los datos en cada iteraci√≥n cambia constantemente el paisaje de p√©rdida (*loss landscape*) para que no se asiente en el m√≠nimo de p√©rdida. Despu√©s de todo, minimizar la p√©rdida es lo que causa el sobreajuste, as√≠ que introducimos un poco de aleatoriedad para crear un poco de subajuste (pero, con suerte, no demasiado).

Por supuesto, nuestra aproximaci√≥n se vuelve m√°s laxa, as√≠ que tenemos que tener cuidado. Por eso hablaremos pronto sobre las divisiones de entrenamiento/prueba, as√≠ como otras m√©tricas para evaluar la fiabilidad de nuestra regresi√≥n lineal.

El ejemplo muestra c√≥mo realizar el descenso de gradiente estoc√°stico en Python. Si cambias el tama√±o de la muestra a m√°s de 1, realizar√° el descenso de gradiente por mini-lotes.

````python
import pandas as pd
import numpy as np

# Datos de entrada
data = pd.read_csv('https://bit.ly/2KF29Bd', header=0)

#Toma todos los valores de X
X = data.iloc[:, 0].values # => el primer parametro ':' quiere decir todas las filas, el segundo quiere decir de la columna 0
print('Values X:\n',X, sep='') # iloc es 'integer location'

#Toma todos los valores de Y
Y = data.iloc[:, 1].values
print('Values Y:\n',Y, sep='')

#Cantidad de filas
n = data.shape[0] 
print('Values n:\n',n, sep='')

# Construyendo el modelo
m = 0.0
b = 0.0

tamanio_muestra = 1  # tama√±o de la muestra
L = .0001  # La tasa de aprendizaje
epochs = 1_000_000  # N√∫mero de iteraciones para realizar descenso de gradiente

# Realizando Descenso de Gradiente Estoc√°stico
for i in range(epochs):
    # choice: primer parametro es el rango de valores, el segundo es el tama√±o de la muestra
    # el tercero indica que los elementos no se pueden repetir
    idx = np.random.choice(n, tamanio_muestra, replace=False)
    print('Index: ',idx)
    x_muestra = X[idx]
    print('X[index]: ',x_muestra)
    y_muestra = Y[idx]
    print('Y[index]: ',y_muestra)

    # El valor actual predicho de Y
    Y_pred = m * x_muestra + b
    

    # Derivada d/dm de la funci√≥n de p√©rdida
    D_m = (-2 / tamanio_muestra) * sum(x_muestra * (y_muestra - Y_pred))

    # Derivada d/db de la funci√≥n de p√©rdida
    D_b = (-2 / tamanio_muestra) * sum(y_muestra - Y_pred)
    m = m - L * D_m  # Actualizar m
    b = b - L * D_b  # Actualizar b

    # imprimir progreso
    if i % 10000 == 0:
        print(i, m, b)

print(f"y = {m}x + {b}")
````

Cuando corr√≠ esto, obtuve una regresi√≥n lineal de *y* = 1.9382830354181135 x + 4.753408787648379  
Obviamente, tus resultados van a ser diferentes, y debido al descenso de gradiente estoc√°stico realmente no vamos a converger hacia un m√≠nimo espec√≠fico, sino que terminaremos en un vecindario m√°s amplio.

### ¬øEs Mala la Aleatoriedad?

Si esta aleatoriedad te resulta inc√≥moda, donde obten√©s una respuesta diferente cada vez que ejecut√°s un fragmento de c√≥digo, ¬°bienvenido al mundo del aprendizaje autom√°tico, optimizaci√≥n y algoritmos estoc√°sticos!  

Muchos algoritmos que hacen aproximaciones se basan en la aleatoriedad, y aunque algunos son extremadamente √∫tiles, otros pueden ser descuidados y tener un rendimiento deficiente, como podr√≠as esperar.

Mucha gente ve el aprendizaje autom√°tico y la IA como alguna herramienta que ofrece respuestas objetivas y precisas, pero eso no podr√≠a estar m√°s lejos de la verdad. ## El Coeficiente de Correlaci√≥n

Mir√° a este diagrama de dispersi√≥n en la Figura junto con su regresi√≥n lineal. ¬øPor qu√© podr√≠a no funcionar demasiado bien una regresi√≥n lineal aqu√≠?  

  


El problema es que los datos tienen una alta varianza. Si los datos est√°n extremadamente dispersos, va a aumentar la varianza hasta el punto de que las predicciones se vuelven menos precisas y √∫tiles, resultando en residuos grandes. Por supuesto, podemos introducir un modelo m√°s sesgado, como la regresi√≥n lineal, para no doblarse y responder a la varianza tan f√°cilmente. Sin embargo, el subajuste tambi√©n va a socavar nuestras predicciones porque los datos est√°n muy dispersos. Necesitamos medir num√©ricamente cu√°nto se "desv√≠an" nuestras predicciones.

Entonces, ¬øc√≥mo mides estos residuos en conjunto? ¬øC√≥mo tambi√©n obtienes una idea de cu√°n mala es la varianza en los datos? Perm√≠teme presentarte el coeficiente de correlaci√≥n, tambi√©n llamado correlaci√≥n de Pearson, que mide la fuerza de la relaci√≥n entre dos variables como un valor entre ‚Äì1 y 1. Un coeficiente de correlaci√≥n m√°s cercano a 0 indica que no hay correlaci√≥n. Un coeficiente de correlaci√≥n m√°s cercano a 1 indica una fuerte correlaci√≥n positiva, lo que significa que cuando una variable aumenta, la otra aumenta proporcionalmente. Si est√° m√°s cerca de ‚Äì1, entonces indica una fuerte correlaci√≥n negativa, lo que significa que a medida que una variable aumenta, la otra disminuye proporcionalmente.

El coeficiente de correlaci√≥n a menudo se denota como *r*. Los datos altamente dispersos en la Figura tienen un coeficiente de correlaci√≥n de 0.1201. Dado que est√° mucho m√°s cerca de 0 que de 1, podemos inferir que los datos tienen poca correlaci√≥n.

Aqu√≠ hay cuatro otros diagramas de dispersi√≥n en la Figura siguiente mostrando sus coeficientes de correlaci√≥n. Nota que cuanto m√°s sigan los puntos una l√≠nea, m√°s fuerte ser√° la correlaci√≥n. Puntos m√°s dispersos resultan en correlaciones m√°s d√©biles.  



Como uno puede imaginarse, el coeficiente de correlaci√≥n es √∫til para ver si hay una posible relaci√≥n entre dos variables. Si hay una relaci√≥n positiva-negativa fuerte, ser√° √∫til en nuestra regresi√≥n lineal. Si no hay una relaci√≥n, pueden agregar solo ruido y perjudicar la precisi√≥n del modelo.  

¬øC√≥mo usamos Python para calcular el coeficiente de correlaci√≥n? Usemos el simple conjunto de datos de 10 puntos que usamos anteriormente. Una manera r√°pida y f√°cil de analizar las correlaciones para todos los pares de variables es utilizando la funci√≥n corr() de Pandas.
````python
import pandas as pd

# Leer datos en un dataframe de Pandas
df = pd.read_csv('https://bit.ly/2KF29Bd', delimiter=",")

# Imprimir correlaciones entre variables
correlations = df.corr(method='pearson')
print(correlations)
````

Como puedes ver, el coeficiente de correlaci√≥n 0.957586 entre *x* e *y* indica una correlaci√≥n positiva fuerte entre las dos variables. Se pueden ignorar las partes de la matriz donde *x* o *y* se calcula en s√≠ mismo (auto-correlaci√≥n) y tiene un valor de 1.0. Obviamente, cuando *x* o *y* se correlaciona en s√≠ mismo, la correlaci√≥n ser√° perfecta en 1.0, porque los valores coinciden exactamente consigo mismos. Cuando tienes m√°s de dos variables, la matriz de correlaci√≥n mostrar√° una cuadr√≠cula m√°s grande porque hay m√°s variables para emparejar y comparar.

Si cambi√°s el c√≥digo para usar un conjunto de datos diferente con mucha varianza, donde los datos est√°n dispersos, vas a ver que el coeficiente de correlaci√≥n disminuye. Esto nuevamente indica una correlaci√≥n m√°s d√©bil.

## Error Est√°ndar de la Estimaci√≥n

Una manera de medir el error general de una regresi√≥n lineal es el SSE, o suma del error cuadrado. Aprendimos sobre esto anteriormente, donde cuadramos cada residual y los sumamos. Si ≈∑ (pronunciado "y sombrero") es cada valor predicho de la l√≠nea e *y* representa cada valor real de *y* de los datos, aqu√≠ est√° el c√°lculo:  

  

Sin embargo, todos estos valores al cuadrado son dif√≠ciles de interpretar, as√≠ que podemos usar algo de l√≥gica de ra√≠z cuadrada para escalar las cosas de vuelta a sus unidades originales. Tambi√©n promediaremos todos ellos, y esto es lo que hace el error est√°ndar de la estimaci√≥n. Si *n* es el n√∫mero de puntos de datos, el ejemplo que viene muestra c√≥mo calculamos el error est√°ndar en Python. 

$$ S_e = \sqrt{\frac{\sum{(y-\hat{y})^2}}{n-2}}

````python
import pandas as pd
from math import sqrt

# Leer los datos
points = list(pd.read_csv('https://bit.ly/2KF29Bd', delimiter=",").itertuples())

n = len(points)

# Regresi√≥n lineal
m = 1.939
b = 4.733

# Calcular el Error Est√°ndar de la Estimaci√≥n
S_e = sqrt((sum((p.y - (m*p.x +b))**2 for p in points))/(n-2))

print(S_e)
````

¬øPor qu√© *n-2* y no *n-1*? Sin entrar demasiado en demostraciones matem√°ticas, se debe a que una regresi√≥n lineal tiene dos variables, no solo una, por lo que debemos aumentar la incertidumbre en uno m√°s en nuestros grados de libertad.

Notar√°s que el error est√°ndar de la estimaci√≥n se parece notablemente a la desviaci√≥n est√°ndar que estudiamos en anteriormente. Esto no es casualidad, ya que es la desviaci√≥n est√°ndar para una regresi√≥n lineal.

## Intervalos de predicci√≥n


Como se mencion√≥ antes, los datos que utilizamos en una regresi√≥n lineal provienen de una muestra de una poblaci√≥n m√°s grande. Esto significa que nuestra regresi√≥n solo refleja la relaci√≥n observada en nuestra muestra espec√≠fica. Adem√°s, la l√≠nea de regresi√≥n que ajustamos tiene una distribuci√≥n normal asociada. En otras palabras, cada valor predicho de *y* tiene su propia distribuci√≥n de probabilidad, similar a c√≥mo la media de una muestra puede variar en diferentes muestras de una poblaci√≥n m√°s grande.

Para simplificar, en una regresi√≥n lineal, buscamos que los datos sigan una distribuci√≥n similar a una campana, con la l√≠nea de regresi√≥n actuando como el punto central o "media". La dispersi√≥n de los datos alrededor de esta l√≠nea nos dice cu√°n dispersos est√°n los datos, de manera similar a c√≥mo la desviaci√≥n est√°ndar mide la dispersi√≥n en una distribuci√≥n de datos. Observemos esto en la siguiente figura:




Cuando tenemos una distribuci√≥n normal siguiendo una l√≠nea de regresi√≥n lineal, no tenemos solo una variable, sino una segunda que tambi√©n influye en la distribuci√≥n. Existe un intervalo de confianza alrededor de cada predicci√≥n de *y*, y esto se conoce como **intervalo de predicci√≥n**.

Volvamos al ejemplo anterior para darle contexto, estimando la edad de un perro y el n√∫mero de visitas al veterinario. Quiero saber el intervalo de predicci√≥n para el n√∫mero de visitas al veterinario con un 95% de confianza para un perro que tiene 8.5 a√±os. Este intervalo de predicci√≥n se muestra en la siguiente figura. Estamos 95% seguros de que un perro de 8.5 a√±os tendr√° entre 16.462 y 25.966 visitas al veterinario.




¬øC√≥mo calculamos esto? Necesitamos obtener el margen de error y sumarlo/restarlo alrededor del valor y predicho. Es una ecuaci√≥n complicada que implica un valor cr√≠tico de la distribuci√≥n T, as√≠ como el error est√°ndar de la estimaci√≥n. Echemos un vistazo:



El valor de x que nos interesa est√° especificado como *xo*, que en el ejemplo es 8.5. A continuaci√≥n, vemos c√≥mo resolverlo en Python:

````python
import pandas as pd
from scipy.stats import t
from math import sqrt

# Leer los datos
points = list(pd.read_csv('https://bit.ly/2KF29Bd', delimiter=",").itertuples())

n = len(points)

# Regresi√≥n Lineal
m = 1.939
b = 4.733

# Calcular el intervalo de predicci√≥n para x = 8.5
x_0 = 8.5
x_mean = sum(p.x for p in points) / len(points)

t_value = t(n - 2).ppf(.975)

standard_error = sqrt(sum((p.y - (m * p.x + b)) ** 2 for p in points) / (n - 2))

margin_of_error = t_value * standard_error * \
                  sqrt(1 + (1 / n) + (n * (x_0 - x_mean) ** 2) / \
                       (n * sum(p.x ** 2 for p in points) - \
                            sum(p.x for p in points) ** 2))

predicted_y = m*x_0 + b

# Calcular el intervalo de predicci√≥n
print(predicted_y - margin_of_error, predicted_y + margin_of_error)
# 16.462516875955465 25.966483124044537
````

Pero, ¬øviste toda la informaci√≥n que obtuvimos hasta el momento?
Creamos una predicci√≥n basada en una regresi√≥n lineal, donde por ejemplo podemos predecir que un perro de 8.5 a√±os tendr√° 21.2145 visitas al veteriario. Pero tambi√©n podemos decir que hay un 95% de probabilidades de que un perro de 8.5 a√±os visite al veterinario entre 16.46 y 25.96 veces. ¬øNo es genial? Es una afirmaci√≥n mucho m√°s segura porque captura un rango en lugar de un valor √∫nico, y as√≠ tiene en cuenta la incertidumbre.

## Divisi√≥n Entrenamiento/Prueba (*Split Train/Test*)

Este an√°lisis detallado a menudo no se lleva a cabo en la pr√°ctica. En ocasiones, los profesionales se enfrentan a tantos datos que no disponen del tiempo ni la capacidad t√©cnica para hacerlo. Por ejemplo, una imagen de 128 √ó 128 p√≠xeles implica al menos 16,384 variables. ¬øSe puede dedicar tiempo a analizar estad√≠sticamente cada una de estas variables de p√≠xeles? Probablemente no. Esto lleva a muchos cient√≠ficos de datos a no tener en cuenta estas m√©tricas estad√≠sticas.

En resumen, cuando se trabaja con grandes vol√∫menes de datos, a menudo es necesario priorizar la escalabilidad sobre la exhaustividad del an√°lisis estad√≠stico. Aunque se pierda en explicabilidad y precisi√≥n, el enfoque de "motosierra" del aprendizaje autom√°tico permite hacer predicciones a gran escala sobre datos masivos. Sin embargo, es importante tener en cuenta que las preocupaciones estad√≠sticas como el sesgo de muestreo y el sobreajuste siguen siendo relevantes, y se deben emplear pr√°cticas de validaci√≥n r√°pida para abordar estas preocupaciones.

Una t√©cnica b√°sica que se utiliza en el aprendizaje autom√°tico para mitigar el sobreajuste (*overfitting*) es una pr√°ctica llamada divisi√≥n de entrenamiento/prueba, donde t√≠picamente se reserva 1/3 de los datos para realizar pruebas y el otro 2/3 se utiliza para el entrenamiento (tambi√©n se pueden usar otras proporciones). El conjunto de datos de entrenamiento se utiliza para ajustar la regresi√≥n lineal, mientras que el conjunto de datos de prueba se utiliza para medir el rendimiento de la regresi√≥n lineal en datos que no ha visto antes. Esta t√©cnica se utiliza generalmente para todo el aprendizaje autom√°tico supervisado, incluyendo la regresi√≥n log√≠stica y las redes neuronales.



````python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Leer los datos
df = pd.read_csv('https://bit.ly/3cIH97A', delimiter=",")

# Extraer las variables de entrada (todas las filas, todas las columnas excepto la √∫ltima)
X = df.values[:, :-1]

# Extraer la columna de salida (todas las filas, la √∫ltima columna)
Y = df.values[:, -1]

# Separar los datos de entrenamiento y prueba
# De esta forma, queda un tercio de los datos para prueba
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=1/3)

model = LinearRegression()
model.fit(X_train, Y_train)
result = model.score(X_test, Y_test)
print("r^2: %.3f" % result)
````

La funci√≥n *train_test_split()* tomar√° nuestro conjunto de datos (columnas X e Y), lo mezclar√° y luego devolver√° nuestros conjuntos de datos de entrenamiento y prueba basados en el tama√±o de nuestro conjunto de datos de prueba. Usamos el m√©todo *fit()* de *LinearRegression* para ajustar los conjuntos de datos de entrenamiento *X_train* e *Y_train*. Luego, usamos el m√©todo *score()* en los conjuntos de datos de prueba X_test e Y_test para evaluar el $R^2$, lo que nos da una idea de c√≥mo se desempe√±a la regresi√≥n en datos que no ha visto antes. Cuanto mayor sea este valor para nuestro conjunto de datos de prueba, mejor ser√°. Tener ese n√∫mero m√°s alto indica que la regresi√≥n se desempe√±a bien en datos que no ha visto antes.

**USANDO R-CUADRADO PARA PRUEBAS**

Veamos el casod e calcular $r^2$ para el caso de una regresi√≥n lineal predefinida a partir del entrenamiento. Estamos comparando los datos de prueba con una l√≠nea de regresi√≥n construida a partir de los datos de entrenamiento. El objetivo sigue siendo el mismo: estar m√°s cerca de 1.0 es deseable para mostrar que la correlaci√≥n de la regresi√≥n es fuerte incluso con los datos de entrenamiento, mientras que estar m√°s cerca de 0.0 indica que el conjunto de datos de prueba tiene un desempe√±o deficiente. 

As√≠ es como se calcula, donde $y_i$ es cada valor y real, $≈∑_i$ es cada valor *y* predicho, y $\bar y$ es el valor *y* promedio para todos los puntos de datos:

$$ r^2 = 1 - \frac{\sum{(y_i-\hat y_i)^2}}{\sum{(y_i-\bar y)^2}} $$

Tambi√©n podemos alternar el conjunto de datos de prueba en cada una de las tres partes. Esto se conoce como validaci√≥n cruzada y a menudo se considera el est√°ndar de oro de las t√©cnicas de validaci√≥n.



El siguiente c√≥digo ejecuta un proceso de validaci√≥n cruzada en tres divisiones del conjunto de datos, evaluando el modelo en cada divisi√≥n de prueba y promediando la m√©trica de puntuaci√≥n (en este caso, el MSE) junto con su desviaci√≥n est√°ndar. Esto nos proporciona una idea de qu√© tan consistentemente se desempe√±a el modelo en diferentes subconjuntos de datos de prueba.

````python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold, cross_val_score

df = pd.read_csv('https://bit.ly/3cIH97A', delimiter=",")

# Extraer las variables de entrada (todas las filas, todas las columnas excepto la √∫ltima)
X = df.values[:, :-1]

# Extraer la columna de salida (todas las filas, la √∫ltima columna)
Y = df.values[:, -1]

# Ejecutar una regresi√≥n lineal simple
kfold = KFold(n_splits=3, random_state=7, shuffle=True)
model = LinearRegression()
results = cross_val_score(model, X, Y, cv=kfold)
print(results)
print("MSE: mean=%.3f (stdev-%.3f)" % (results.mean(), results.std()))
````

Cuando est√°s preocupado por la variabilidad en tu modelo, una t√©cnica √∫til es realizar una validaci√≥n cruzada con pliegues aleatorios. Esto implica dividir repetidamente tus datos en conjuntos de entrenamiento y prueba de forma aleatoria y promediar los resultados de las pruebas para evaluar la consistencia del rendimiento del modelo. En el siguiente ejemplo, se realiza este proceso 10 veces, calculando luego el promedio de los resultados de las pruebas y sus desviaciones est√°ndar.

¬øLa desventaja? Es computacionalmente muy costoso, ya que estamos entrenando la regresi√≥n muchas veces.

````python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score, ShuffleSplit

df = pd.read_csv('https://bit.ly/38XwbeB', delimiter=",")

# Extraer las variables de entrada (todas las filas, todas las columnas excepto la √∫ltima)
X = df.values[:, :-1]

# Extraer la columna de salida (todas las filas, la √∫ltima columna)
Y = df.values[:, -1]

# Ejecutar una regresi√≥n lineal simple
kfold = ShuffleSplit(n_splits=10, test_size=.33, random_state=7)
model = LinearRegression()
results = cross_val_score(model, X, Y, cv=kfold)

print(results)
print("mean=%.3f (stdev-%.3f)" % (results.mean(), results.std()))
````

Entonces, cuando tienes poco tiempo o tus datos son demasiado voluminosos para analizar estad√≠sticamente, una divisi√≥n de entrenamiento/prueba proporcionar√° una forma de medir qu√© tan bien se desempe√±ar√° tu regresi√≥n lineal en datos que no ha visto antes.

Es importante tener en cuenta que simplemente aplicar las mejores pr√°cticas de aprendizaje autom√°tico, como dividir tus datos de entrenamiento y prueba, no garantiza que tu modelo vaya a funcionar bien. Puedes ajustar demasiado tu modelo y obtener buenos resultados en las pruebas de forma artificial, solo para descubrir que no funciona bien en el mundo real. Por eso, a veces es necesario mantener otro conjunto de datos llamado **conjunto de validaci√≥n**, especialmente si est√°s comparando diferentes modelos o configuraciones. De esta manera, los ajustes que realices en los datos de entrenamiento para obtener un mejor rendimiento en los datos de prueba no filtrar√°n informaci√≥n al modelo durante el entrenamiento. Puedes usar el conjunto de validaci√≥n como una √∫ltima medida para ver si el ajuste caus√≥ un sobreajuste a los datos de prueba.

Incluso as√≠, es posible que todo tu conjunto de datos (incluidos entrenamiento, prueba y validaci√≥n) est√© sesgado desde el principio, y ninguna divisi√≥n podr√° mitigar eso.
